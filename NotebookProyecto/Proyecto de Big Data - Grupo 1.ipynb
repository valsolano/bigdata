##  Antecedentes

En el mundo del consumismo en el que vivimos, la industria aérea representa uno de los sectores más dinámicos en términos de fijación de presios ya que, a diferencia de otros productos que podemos encontrar en el mercado con precios más estables, los tiquetes de avión suelen experimentar cambios significativos en cuestión de segundos. Estas inconsistencias pueden depender de múltiples factores como la demanda, comptencia, costos operativos, y algunas más específicas para cada vuelo como la duración, si cuenta con escalas o la clase del servicio.
En el mercado aéreo de la India, particularmente en las rutas que conectan las seis principales ciudades metropolitanas (Delhi, Mumbai, Bangalore, Kolkata, Hyderabad y Chennai), se ha experimentado un crecimiento en la demanda de viajes aéreos. Las agencias de viajes en línea operan en un entorno altamente competitivo donde deben ofrecer a sus clientes las mejores recomendaciones de compra para mantener su lealtad y confianza. Por lo tanto, el desarrollo de un modelo que pueda analizar patrones de opciones de vuelo y predecir con precisión los precios futuros se vuelve una necesidad estratégica para mejorar la experiencia del cliente y mantenerse competitivos en el mercado.


## Definición del problema

El precio de los vuelos aéreos presenta una alta variabilidad debido a factores como la aerolínea, el origen y destino, la duración del vuelo, el número de escalas, la fecha de viaje y el momento de compra, lo que dificulta que los usuarios puedan tomar decisiones informadas al adquirir boletos aéreos. Aunque existe una gran cantidad de datos históricos disponibles, estos no siempre son aprovechados mediante técnicas de Big Data y análisis predictivo para anticipar el comportamiento de los precios. Por ello, surge la necesidad de desarrollar un modelo de predicción de precios de vuelos basado en el análisis de grandes volúmenes de datos, que permita estimar de manera más precisa el costo de los boletos y apoyar la toma de decisiones eficientes.


## Justificación 

Implementar un modelo de aprendizaje automático para predecir el comportamiento de los precios de los boletos aéreos facilitará el análisis automatizado de grandes volúmenes de datos provenientes de las estimaciones. Este enfoque permitirá comprender variables como la demanda, la temporada, la ruta y el tiempo de anticipación influyen en las tarifas, apoyando la toma de decisiones estratégicas relacionadas con la gestión de precios. Además, el uso de modelos predictivos contribuirá a detectar patrones de mercado relevantes, identificar oportunidades en rutas específicas y fortalecer las estrategias comerciales, generando ventaja competitiva y un mayor valor tanto para la empresa como para los pasajeros. 

## Objetivo general

Desarrollar un modelo de predicción del precio de vuelos mediante el análisis de grandes volúmenes de datos históricos, utilizando técnicas de Big Data y aprendizaje automático, con el fin de estimar de manera precisa el costo de los boletos y apoyar la toma de decisiones informadas para los usuarios.

### Objetivos específicos

## Descripción del conjunto de datos

Para este trabajo se utiliza como fuente el dataset de Kaggle “Flight Price Prediction”. Este dataset trata sobre precios de tiquetes de avión y opciones de reserva de vuelos. La información fue obtenida originalmente del sitio EaseMyTrip, que es una plataforma en línea donde las personas compran vuelos, y fue recolectada mediante web scraping utilizando la herramienta Octoparse.

En el dataset, cada fila representa una opción distinta de vuelo e incluye datos como la aerolínea, el código de vuelo, la ciudad de salida y destino, la hora de salida y llegada (agrupadas en rangos), la cantidad de escalas, la duración del viaje, los días restantes entre la compra y el vuelo, la clase del tiquete (Economy o Business) y el precio, que es la variable principal que se quiere analizar y predecir.

La idea de usar este dataset en el proyecto es poder estudiar qué factores influyen en el precio y luego entrenar modelos para predecir el precio según las características del vuelo.

### Características del Dataset

El dataset consta de las siguientes columnas:  

* Airline: almacena el nombre de la aerolínea. Es una variable categórica con 6 aerolíneas distintas.
* Flight: guarda información del código del vuelo. Es una variable categórica.
* Source City: indica la ciudad de salida del vuelo. Es una variable categórica con 6 ciudades únicas.
* Departure Time: indica la hora de salida, pero está agrupada en rangos. Es una variable categórica derivada con 6 etiquetas de tiempo.
* Stops: representa la cantidad de escalas entre origen y destino. Es una variable categórica con 3 valores distintos.
* Arrival Time: indica la hora de llegada, también agrupada en rangos. Es una variable categórica derivada con 6 etiquetas de tiempo.
* Destination City: indica la ciudad de destino donde aterriza el vuelo. Es una variable categórica con 6 ciudades únicas.
* Class: indica la clase del tiquete. Es una variable categórica con 2 valores: Business y Economy.
* Duration: muestra la duración total del viaje en horas. Es una variable continua.
* Days Left: representa los días restantes entre la fecha de compra y la fecha del vuelo (se calcula como fecha del viaje - fecha de reserva). Es una variable derivada.
* Price: es la variable objetivo, y almacena el precio del tiquete.

### Tamaño del Dataset

* Instancias: 300,261 registros.
* Atributos: 11 características (10 predictoras y 1 variable objetivo: Price).

## Objetivo propuesto

* Construir modelos que predigan el precio del tiquete de avión con base en las características del vuelo disponibles en el dataset.

### Limitaciones

* Solo tiene vuelos entre 6 ciudades principales de India, entonces no aplica para otros destinos.
* Los datos son de un rango de fechas específico (11 de febrero a 31 de marzo de 2022), así que en otras temporadas los precios pueden ser diferentes.
* La hora de salida y llegada están agrupadas por rangos y no en horas exactas, lo que reduce el detalle del análisis.
* Al venir de web scraping de una página (EaseMyTrip), los precios pueden depender de lo que el sitio mostraba en ese momento.
